```python
def myfunc():
    x = ''
    while True:
        print(f'Yielding x ({x}) and waiting')
        x = yield x
        if x is None: 
            break
        print(f'Got x {x}. Doubling.')
        x = x * 2

g = myfunc() 
next(g)
g.send(10)
g.send(123)
```
