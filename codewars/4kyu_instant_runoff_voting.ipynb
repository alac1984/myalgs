```python
# /kata/52996b5c99fdcb5f20000004/train/python
# This was FUCKING HARD, but I DID IT ALMOST ALONE! HELL YEAAAAAHHH!

from collections import defaultdict

def runoff(voters: list) -> str:
    count = defaultdict(int)
    # Create the list of candidates
    for candidate in voters[0]:
        count[candidate] = 0
        
    # Count the votes for first choice
    for voter in voters:
        count[voter[0]] += 1
    
    # from IPython.core.debugger import Pdb; Pdb().set_trace()
    # Check if there's just one candidate left
    if len(count.keys()) == 1:
        return voters[0][0]
    
    # Check for the complete tie possibility
    vote_count = set(count.values())
    if len(vote_count) == 1:
        return None
    
    # Initialize conditions
    max_votes = 0
    min_votes = 99999999999
    most_voted_group = []
    most_voted_draw = False
    least_voted_draw = False
    least_voted_group = []
    # Start counting
    
    for candidate, votes in count.items():
        # Check if is the highest vote score so far
        if votes > max_votes:
            max_votes = votes
            most_voted_draw = False
            most_voted_group = [candidate]
        # Check for draw on most voted
        if votes == max_votes and candidate not in most_voted_group:
            most_voted_draw = True
            most_voted_group.append(candidate)
        
        # Check if is the lowest vote score so far
        if votes < min_votes:
            min_votes = votes
            least_voted_draw = False
            least_voted_group = [candidate]
        # Check for draw on least voted
        if votes == min_votes and candidate not in least_voted_group:
            least_voted_draw = True
            least_voted_group.append(candidate)
            
    # If a candidate has more than half the votes, he wins
    if len(most_voted_group) == 1 and len(least_voted_group) == 0:
        return most_voted_group[0]
    else:
        # Remove least voted candidate from ballots
        new_voters = []
        for voter in voters:
            new_voter = [candidate for candidate in voter if candidate not in least_voted_group]
            new_voters.append(new_voter)
        # Run the function with the new ballots
        return runoff(new_voters)
        
```

```python
# Best Codewars answer

from collections import Counter

def runoff(voters):
    from IPython.core.debugger import Pdb; Pdb().set_trace()
    while voters[0]:
        poll = Counter(ballot[0] for ballot in voters)
        winner, maxscore = max(poll.items(), key = lambda x: x[1])
        minscore = min(poll.values())
        if maxscore * 2 > len(voters):
            return winner
        voters = [[c for c in voter if poll[c] > minscore] for voter in voters]
```
