A chess algorithm runs in O(1) time. It is, however, too big, but bear in mind 
that chess has a large but finite number of positions.

```
def chess_analyzer(game: ChessGame):
    input: an object that abstracts a chess game
    output: the victorious player after all perfect moves
    
    while not board.is_checkmate():
        player = board.next_player
        player.make_perfect_move()
       
    return game.winner
```

`Board`: a class that abstracts a chess game (with players, board, and methods to determine winner or the next player.

`Board.next_player`: a property whose value is updated every a perfect_move() is made. Runs in O(1) time

`Board.is_checkmate`: a method that determines if the king of `Board.next_player` has any legal moves. Runs on O(1) time.

`Board.make_perfect_move`: a method that calculates all possible combination of moves that led to a checkmate of opposite king. It creates a tree that covers all opponent's possible answers and execute only the moves that led to its checkmate. It runs in O(1) time, but the size of this calculation is pretty huge!

````python
class Printer:
    count: int = 0
    
    def __init__(self):
        self.string = ""
        
    def show(self):
        self.count += 1
        print(self.string)
    
    def put(self, string: str):
        self.string = string
    
    
def my_own_song(n: int, printer: Printer):
    if n == 0:
        printer.put("Goodbye!")
        printer.show() # 1
        print(printer.count)
        return
        
    printer.put(f"I wanna kiss you...")
    printer.show() # n
    
    for i in reversed(range(1, n+1)):
        printer.put(f"    {i} times") 
        printer.show() # n + (n-1) + (n-2) ... + (n-n)
    my_own_song(n-1, printer)

# For n = 1, 3 operations
# For n = 2, 6 operations, Δ = 3
# For n = 3, 10 operations, Δ = 4
# For n = 4, 15 operations, Δ = 5
# For n = 5, 21 operations, Δ = 6
# For n = 6, 28 operations, Δ = 7
# For n = 7, 36 operations, Δ = 8
# Could not determine its running time!!!!!!!!!!!!!!!!!!!
printer = Printer()
my_own_song(7, printer)
        
        
````

````python
def n_days_of_xmas(gifts: list[str]) -> None:
    # import ipdb; ipdb.set_trace()
    for n in range(1, len(gifts)+1):
        print(f"On the {n} day of Christmas, my true love gave to me") # n
        for idx in range(1, n): # n
            print(f"{idx+1} {gifts[idx]}")
        if n > 1:
            print("and")
        print("a partridge in a pear tree\n")

# O(n²)
````

````python
gifts = [
    "balls",
    "sack",
    "eggs",
    "salsage"
]
n_days_of_xmas(gifts)
````
