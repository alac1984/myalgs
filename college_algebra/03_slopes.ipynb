```python
# The "slope" is the m in y = mx + b. It represents the steepness of a linear function.
# Given two points, we can get the slope of the function that passes over them like this:

x1 = 1
y1 = 7
x2 = 2
y2 = 10

# The slope is:
m = (y2 - y1) / (x2 - x1)
m
```

```python
# With the slope, we can write the equation

b = y1 - m*x1
print("y = ", int(m), "x + ", int(b))
```

```python
# Plotting
import sympy as sp
import matplotlib.pyplot as plt

fig, ax = ptl.subplots()
plt.axis([-20, 20, -20, 20])
plt.plot([-20, 20], [0, 0], 'b')
plt.plot([0, 0], [-20, 20], 'b')

# Getting the x of the equation
x = sp.Symbol('x')
y = 3*x + 4
domain_set = range(-20, 20)
image_set = []

for domain_val in domain_set:
    image_val = y.subs(x, domain_val)
    image_set.append(image_val)

plt.plot(domain_set, image_set, color="red", label="y = 3x + 4")
plt.show()
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
plt.axis([2010, 2020, 0, 150000])

# Getting the x of the equation
x = sp.Symbol('x')
y = 10500*x - 21050000
domain_set = range(2010, 2020, 2)
image_set = []

for domain_val in domain_set:
    image_val = y.subs(x, domain_val)
    image_set.append(image_val)
    
ax.set_xticks(np.arange(2010, 2020, 3))
ax.set_yticks(np.arange(0, 150000, 20000))

plt.plot(domain_set, image_set, 'o--', color="red", label="Population growth")
plt.legend()
plt.show()
```
