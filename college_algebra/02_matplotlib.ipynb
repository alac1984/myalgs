```python
# Plotting a point

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.axis([-10, 10, -10, 10]) # axis size
plt.plot([-10, 10], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [-10, 10], 'b') # blue y axis
plt.plot([5], [4], 'ro')
plt.show()
```

```python
# Plotting multiple points

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.axis([-10, 10, -10, 10]) # axis size
plt.plot([-10, 10], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [-10, 10], 'b') # blue y axis
print("x \t y")
for x in range (-10, 11):
    y = 0.5*x + 1
    plt.plot([x], [y], 'ro')
    print(x, "\t", y)
plt.show()
```

```python
# Use arrays as input

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 40)

fig, ax = plt.subplots()
plt.axis([-10, 10, -10, 10]) # axis size
plt.plot([-10, 10], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [-10, 10], 'b') # blue y axis
y = 2*x + 1
plt.plot(x, y, "r")

plt.show()
```

```python
# Customizing a bunch more

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 40)

fig, ax = plt.subplots()
plt.axis([-12, 12, -12, 12]) # axis size
plt.plot([-12, 12], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [-12, 12], 'b') # blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Some graph")
ax.set_xticks(np.arange(-10, 12, 1))
ax.set_yticks(np.arange(-10, 12, 1))
ax.grid(True)

y = 2*x + 1
z = -3*x - 10
plt.plot(x, y, label='y = 2x + 1')
plt.plot(x, z, label='z = -3x - 10')
plt.legend()
plt.show()
```
