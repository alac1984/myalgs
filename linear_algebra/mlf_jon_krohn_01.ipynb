```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot

jtplot.style(theme='oceans16')
```

```python
t = np.linspace(0, 40, 1000)
t
```

<br>
Distance travelled by the robber: $d = 2.5t$
<br>
<br>


```python
d_r = 2.5 * t
d_r
```

<br>
Distance travelled by the sheriff: $d = 3(t - 5)$
<br>
<br>

```python
d_s = 3 * (t - 5)
d_s
```

```python
fig, ax = plt.subplots()
plt.title("A Bank Robber Caught")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c="green")
ax.plot(t, d_s, c="brown")
plt.axvline(x=30, color="purple", linestyle="--")
_ = plt.axhline(y=75, color="purple", linestyle="--")
```

```python
t = np.linspace(0, 60, 1000)
m1 = t
m2 = 4 * (t - 30)

fig, ax = plt.subplots()
plt.title("Solar Panels Energy Generation")
plt.xlabel("time (in days)")
plt.ylabel("energy (in kJ)")
ax.set_xlim([0, 60])
ax.set_ylim([0, 200])
ax.plot(t, m1, c="green")
ax.plot(t, m2, c="brown")
plt.axvline(x=40, color="purple", linestyle="--")
_ = plt.axhline(y=40, color="purple", linestyle="--")
```
