```python
class Disc:
    def __init__(self, name: str, size: int):
        self.name = name
        self.size = size
    
    def __repr__(self) -> str:
        string = "[]"
        for _ in range(self.size):
            middle = len(string) // 2
            string = string[:middle] + "  " + string[middle:]
            
        middle = len(string) // 2
        string = string[:middle] + self.name + string[middle:]
        
        if len(string) < 11:
            diff = 11 - len(string)
            # import ipdb; ipdb.set_trace()
            for _ in range(diff):
                string += " "
        
        return string
    
class Rod(list):
    def __init__(self, name: str):
        super().__init__()
        self.name = name
        
    def append(self, disc: Disc) -> None:
        try:
            top = self[0]
        except:
            super().append(disc)
            return
            
        if disc.size > top.size:
            raise ValueError("You cannot put a greater disk on top of a smaller one")
        
        super().append(disc)
    
    def __repr__(self) -> str:
        string = "-----------\n"
        rev = reversed(self)
        for el in rev:
            string += str(el) + "\n"
        return string + "-----------"

class Board:
    def __init__(self, ndisc: int, nrod: int):
        self.ndisc = ndisc
        self.nrod = nrod
        self.rods = [Rod(f"{n}") for n in range(nrod)]
        discs = [Disc(f"{n}", n) for n in range(ndisc)]
        discs.reverse()
        # import ipdb; ipdb.set_trace()
        for disc in discs:
            self.rods[0].append(disc)
    
    def __repr__(self):
        # Split lines from rod strings
        rod_strings = []
        for rod in self.rods:
            lines = rod.__repr__().split("\n")
            for line in lines:
                # import ipdb; ipdb.set_trace()
                if len(line) < 11:
                    diff = 11 - len(line)
                    for _ in range(diff):
                        line += " "
            rod_strings.append(lines)
        
        # Give rods the size to receive the all discs
        for rod_string in rod_strings:
            if len(rod_string) < self.ndisc + 2:
                diff = self.ndisc - len(rod_string) + 2
                for _ in range(diff):
                    rod_string.insert(1, "           ") # 11 spaces :)
                        
        # Creating the string
        string = ""
        for i in range(self.ndisc + 2):
            for idx, rod_string in enumerate(rod_strings):
                # import ipdb; ipdb.set_trace()
                if idx == self.nrod - 1:
                    string += rod_string[i] + "   "
                    string += "\n"
                else:
                    string += rod_string[i] + "   "
            
        return string
    
    def move(self, src: Rod, dst: Rod) -> None:
        # import ipdb; ipdb.set_trace()
        disc = src.pop()
        dst.append(disc)
        print(self.__repr__())
        
```
