```python
"""Imports"""
import sympy as sy
import matplotlib.pyplot as plt
from sympy.core.relational import Equality
from sympy.core.symbol import Symbol
```

```python
"""Function that solve an equation for multiple values"""

def eq_solutions(eq: Equality, symbol: Symbol, values: list[int]) -> list:
    # List equation symbols
    symbols = list(eq.free_symbols)
    # Create a dict with all symbols
    subs = {s: s if s != symbol else None for s in symbols}
    solutions = []
    
    for v in values:
        # Change the value of the symbol to substitute
        subs[symbol] = v
        # Substitute the equation for the new one with values
        subs_eq = eq.subs(subs)
        solutions.append(sy.solve(subs_eq)[0])
    
    return solutions
    
```

```python
"""Big O Examples, page 23"""

# Example 1
sy.var('y n')
eq = sy.Eq(y, 3*n + 8)
eq_nval = list(range(1, 30))
eq_yval = eq_solutions(eq, n, eq_nval)

sy.var('y n')
bigo = sy.Eq(y, 4*n)
bigo_nval = list(range(1, 30))
bigo_yval = eq_solutions(bigo, n, bigo_nval)

plt.plot(eq_nval, eq_yval, color="green", linestyle="dashed", label="eq")
plt.plot(bigo_nval, bigo_yval, color="blue", linestyle="dashed", label="big_o")
plt.legend()
plt.title("Example 1")
plt.show()
```
